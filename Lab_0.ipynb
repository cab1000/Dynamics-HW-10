```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

x = np.array([20,100,0,-80,-210,50,75])
y= [15.1, 18.6, 13.64, 10, 3.84, 16.13, 17.38]

def func(x, a, b):
  y = a*x + b
  return y

alpha = curve_fit(func, xdata = x, ydata = y)[0]
print(alpha)

plt.scatter(x,y)
plt.plot(x, alpha[0]*x + alpha[1], 'r')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import chi2

x = np.array([20,100,0,-80,-210,50,75])
y= np.array([15.1, 18.6, 13.64, 10, 3.84, 16.13, 17.38])
e= np.array([5,5,4,10,20,2,6])

F = np.polyfit(x,y, 1, w = e )
chi = np.sum((np.polyval(F, x) - y) ** 2)

plt.plot(x, chi2.pdf(x, df=4))
```
